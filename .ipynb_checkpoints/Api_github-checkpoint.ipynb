Esquema da API do github: https://api.github.com/  
Documentação da api: https://developer.github.com/v3/  
Documentação do request: https://requests.readthedocs.io/pt_BR/latest/index.html

```python
import requests
import json
```

```python
user_input = input('Digite o usuario github para a consulta: ')
```

#### Dados do user

```python
request_user = requests.get('https://api.github.com/users/{}'.format(user_input))
```

```python
address_data_user = request_user.json()
address_data_user
# Aqui a gente ve a quantidade de informaçoes que podemos pegar
```

```python
address_data_user.items()
```

```python
for k, v in address_data_user.items():
    print(k,":", v)
```

```python
address_data_user['name']
```

#### Dados dos repositorios

```python
request_repo = requests.get('https://api.github.com/users/{}/repos'.format(user_input))
```

```python
address_data_repo = request_repo.json()
address_data_repo
```

```python
len(address_data_repo)
```

```python
address_data_repo[1]['name']
```

```python
for i in range(len(address_data_repo)):
    print(address_data_repo[i]['name'])
```

```python
for i in range(len(address_data_repo)):
    print(address_data_repo[i]['url'])
```
